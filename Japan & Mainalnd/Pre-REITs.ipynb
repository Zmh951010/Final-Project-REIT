```python
# -*- coding: utf-8 -*-
import pandas as pd
from pyecharts import Scatter
```

```python
df = pd.read_csv('RCREITS.csv',encoding="gb2312")
```

```python
type(df['time'][9])
```

```python
project = []
for i in range(0,len(df['name'])):
    Project = df['name'][i].replace("?",'')
    project.append(Project)
```

```python
df['name'] = project
```

```python
df
```

```python
df['date'] = pd.to_datetime(df['time'])
```

```python
df
```

```python
scatter = Scatter("Mainland Pre-REITs Projects", width=800, height=600,title_top = 20)

scatter.add("Commercial",
           df['date'][df['Sector'] == 'Commercial'],
           df['Issuing_Value(CNY_Million)'][df['Sector'] == 'Commercial'],
           symbol_size = 15
           )

scatter.add("Hotels",
           df['date'][df['Sector'] == 'Hotels'],
           df['Issuing_Value(CNY_Million)'][df['Sector'] == 'Hotels'],
           symbol_size = 15
           )
scatter.add("Office",
           df['date'][df['Sector'] == 'Office'],
           df['Issuing_Value(CNY_Million)'][df['Sector'] == 'Office'],
           symbol_size = 15
           )
scatter.add("Residential",
           df['date'][df['Sector'] == 'Residential'],
           df['Issuing_Value(CNY_Million)'][df['Sector'] == 'Residential'],
           symbol_size = 15
           )
scatter.add("Warehousing",
           df['date'][df['Sector'] == 'Warehousing'],
           df['Issuing_Value(CNY_Million)'][df['Sector'] == 'Warehousing'],
           symbol_size = 15,
           xaxis_name="Issue Date",
           xaxis_type = "time",
           yaxis_name="Issuing Value (CNY Million)",
           yaxis_name_gap=50,
           xaxis_name_gap=40
            
           )
scatter.show_config()
scatter.render('Pre-REITscat.html')
```

```python
df1 = pd.read_csv('Pre-REITs-proportion.csv')
```

```python
df1
```

```python
from pyecharts import Pie
```

```python
attr = df1['Sector']
v1 = df1['Proportion']
pie = Pie("Pre-REITs Proportion", width=600, height=500)
pie.add(
    "",
    attr,
    v1,
    radius=[20, 50],
    label_text_color=None,
    is_label_show=True,
    legend_orient="vertical",
    legend_top = "center",
    legend_pos="left",
    label_pos = "inside",
    label_formatter='{d}%'
)
pie.render('Pre-REITpie.html')
```
