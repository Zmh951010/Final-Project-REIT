```python
from pyecharts import Scatter
```

```python
import pandas as pd
```

```python
dfj = pd.read_csv('J-REITs_Sector.csv')
```

```python
dfj
```

```python
cap=[]
for i in dfj['caps']:
    newcap=int(i.replace(",",""))
    cap.append(newcap)
```

```python
dfj['caps']=cap
```

```python
dfj['code']=dfj['code'].astype('str')
scatter = Scatter("J-REITs Overview", width=1900, height=1200,title_top = 20)
scatter.add("Hotels", 
            dfj['yield'][dfj['sector'] == 'Hotels'], 
            dfj['caps'][dfj['sector'] == 'Hotels'],
            symbol_size = 30
            
            )
scatter.add("Office", 
            dfj['yield'][dfj['sector'] == 'Office'], 
            dfj['caps'][dfj['sector'] == 'Office'],
            
            )
scatter.add("Logisitics", 
            dfj['yield'][dfj['sector'] == 'Logistics'], 
            dfj['caps'][dfj['sector'] == 'Logistics'], 
            
            )

scatter.add("Residential", 
            dfj['yield'][dfj['sector'] == 'Residential'], 
            dfj['caps'][dfj['sector'] == 'Residential'],
            
           
            )
scatter.add("Commercial", 
            dfj['yield'][dfj['sector'] == 'Commersial'], 
            dfj['caps'][dfj['sector'] == 'Commersial'],
            
            )
scatter.add("Diversified", 
            dfj['yield'][dfj['sector'] == 'Diversified'], 
            dfj['caps'][dfj['sector'] == 'Diversified']
      
            )
scatter.add("Complex", 
            dfj['yield'][dfj['sector'] == 'Complex'], 
            dfj['caps'][dfj['sector'] == 'Complex'],
            
           )
scatter.add("Healthcare", 
            dfj['yield'][dfj['sector'] == 'Healthcare'], 
            dfj['caps'][dfj['sector'] == 'Healthcare'],
            
            xaxis_name="Dividend Yield(%)",
            yaxis_name="Market Cap(million yen)",
            xaxis_min=0.02,
            yaxis_name_gap=60,
            legend_top = 'bottom'
           )
scatter.show_config()
scatter.render('JREITscat.html')
```
