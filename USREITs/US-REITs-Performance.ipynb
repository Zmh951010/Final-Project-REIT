```python
import pandas as pd
```

```python
df = pd.read_csv('US_REITS_Performance.csv')
```

```python
df1 = df.dropna()
```

```python
df1
```

```python
NAV = []
for i in range(0,len(df1['Net_Assets'])):
    newNAV = float(df['Net_Assets'][i].replace("M",''))    
    NAV.append(newNAV)
```

```python
df1['NAV_in_million'] = NAV
```

```python
df2 = df1.drop('Net_Assets',1)
```

```python
df2['CapGain'] = df2['YTD_Return'] - df2['Yield']
```

```python
df2
```

```python
from pyecharts import Scatter
```

```python
Y = []
for i in range(0,len(df2['Yield'])):
    Yd = df2['Yield'][i]  
    Y.append(Yd)
CG = []
for i in range(0,len(df2['CapGain'])):
    cg = df2['CapGain'][i]  
    CG.append(cg)
extra_data1 = []
for i in range(0,len(df2['NAV_in_million'])):
    nav = df2['NAV_in_million'][i]  
    extra_data1.append(nav)
scatter = Scatter("US-REITs Performance(size = Net Assets Value in million)",title_text_size  = 50, width=1900, height=1200,title_top = 20)
scatter.add("US REITs", 
            Y, 
            CG,
            extra_data=extra_data1,
            is_visualmap=True,
            visual_dimension=2,
            visual_orient="horizontal",
            visual_type="size",
            visual_range=[2.82, 59160],
            visual_range_size = [5,50],
            xaxis_name="Dividend Yield(%)",
            yaxis_name="Capital Gain(%)",
            xaxis_name_size = 24
           )
scatter.render('US-REITscat.html')
```
