```python
import pandas as pd
from pyecharts import Line
from pyecharts import Bar
import matplotlib.pyplot as plt
import seaborn
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from pyecharts import HeatMap
```

```python
df = pd.read_csv('HKreits.csv')
```

```python
df
```

```python
line = Line("Different Investment Comparison (Hong Kong , 2008-2018) ",width=1000, height=600)
line.add("REITs", df['Time'], df['REITS Monthly Return(%)'], mark_line=["average"])
line.add("HIS Index", df["Time"], df['HIS Index'], mark_line=["average"])
line.add("CPI",df["Time"], df['CPI (%) '], mark_line=["average"])
line.add("HKMA Policy Rate", df["Time"], df['Hong Kong Monetary Authority Policy Rate(%)'], mark_line=["average"])
line.add("Bond", df["Time"], df['Ten-year Government Bond Interest Rate(%)'], mark_line=["average"])
line.add("House Price", df["Time"], df['House price'], mark_line=["average"],
         legend_orient='vertical',
         is_datazoom_show=True,
         datazoom_type='slider',
         legend_top='top',
         legend_pos='right'
        
        )
line.render('lines.html')
```

```python
tenyears=list(df.loc[:,'REITS Monthly Return(%)':'House price'].sum())
```

```python
fiveyears=list(df.loc[69:129, 'REITS Monthly Return(%)':'House price'].sum())
```

```python
oneyear=list(df.loc[117:129, 'REITS Monthly Return(%)':'House price'].sum())
```

```python
attr=['REITs', 'HKMA Policy Rate','Bond', 'CPI','HIS Index','House Price']
```

```python
bar = Bar("10-Year/5-Year/1-Year Accumulated  Rate Comparison (HK, 2008-2018)", width=1200, height=600)
bar.add("10-Year", attr, tenyears)
bar.add("5-Year", attr, fiveyears)
bar.add("1-Year", attr, oneyear,
        legend_orient='vertical',
        
        legend_top='10%',
        legend_pos='right')
bar.render('bar.html')
```

```python
def plot_corr(df,size=7):
    '''Function plots a graphical correlation matrix for each pair of columns in the dataframe.

    Input:
        df: pandas DataFrame
        size: vertical and horizontal size of the plot'''

    corr = df.corr()
    fig, ax = plt.subplots(figsize=(size, size))
    ax.matshow(corr)
    plt.xticks(range(len(corr.columns)), corr.columns)
    plt.yticks(range(len(corr.columns)), corr.columns)
```

```python
plot_corr(df)
```

```python
x_axis = ['REITS', 'HKMA Policy Rate', 'CPI ', 'Bond',
       'HIS Index', 'House price']
y_axis =  ['REITS', 'HKMA Policy Rate', 'CPI ', 'Bond',
       'HIS Index', 'House price']
```

```python
from pyecharts import HeatMap
#data = [[i, j, random.randint(0, 50)] for i in range(24) for j in range(7)]
heatmap = HeatMap('Correlation Matrix of Different Type of Investment', width=1200, height=1050)
heatmap.add("Correlation among products",
    x_axis,
    y_axis,
    data,
    xaxis_interval=0,
    xaxis_pos='top',
    yaxis_interval=0,
    is_yaxis_inverse=True,
    is_label_show=True,
    is_visualmap=True,
    visual_text_color="#000",
    visual_range_color=['#1E90FF','#FFFFFF','#FF0000'],
    visual_range=[-1, 1],
    visual_pos='right',
    visual_top='top',
    is_legend_show=True
           )
heatmap.render("HKcorr.html")
```
